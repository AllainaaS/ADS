```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

skewness_temp = dataset['Temperature'].skew()
skewness_humidity = dataset['Humidity'].skew()
skewness_wind_speed = dataset['Wind Speed'].skew()

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(dataset['Temperature'], kde=True)
plt.title(f'Temperature Distribution (Skewness: {skewness_temp:.2f})')
plt.xlabel('Temperature')

plt.subplot(1, 3, 2)
sns.histplot(dataset['Humidity'], kde=True)
plt.title(f'Humidity Distribution (Skewness: {skewness_humidity:.2f})')
plt.xlabel('Humidity')

plt.subplot(1, 3, 3)
sns.histplot(dataset['Wind Speed'], kde=True)
plt.title(f'Wind Speed Distribution (Skewness: {skewness_wind_speed:.2f})')
plt.xlabel('Wind Speed')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import kurtosis

kurtosis_temp = kurtosis(dataset['Temperature'])
kurtosis_humidity = kurtosis(dataset['Humidity'])
kurtosis_wind_speed = kurtosis(dataset['Wind Speed'])

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.histplot(dataset['Temperature'], kde=True)
plt.title(f'Temperature Distribution (Kurtosis: {kurtosis_temp:.2f})')
plt.xlabel('Temperature')

plt.subplot(1, 3, 2)
sns.histplot(dataset['Humidity'], kde=True)
plt.title(f'Humidity Distribution (Kurtosis: {kurtosis_humidity:.2f})')
plt.xlabel('Humidity')

plt.subplot(1, 3, 3)
sns.histplot(dataset['Wind Speed'], kde=True)
plt.title(f'Wind Speed Distribution (Kurtosis: {kurtosis_wind_speed:.2f})')
plt.xlabel('Wind Speed')

plt.tight_layout()
plt.show()
```
